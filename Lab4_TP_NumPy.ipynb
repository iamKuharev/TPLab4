```python
import numpy as np
```

Создание матрицы М:

image.png

```python
#Создание заданной матрицы, заполненной цифрой 2
dimension = 85
m = np.full((dimension, dimension), 2)

#Создание главной диагоняли
diagonal = np.arange(3, 3 + 2 * dimension, 2)
np.fill_diagonal(m, diagonal)

m
```

Создание матрицы N:

image.png

```python
#Создание заданной матрицы, заполненной цифрой 4
n = np.full((dimension, dimension), 4)

#Замена символов
n[::2, 2::3] = 5
n[1::2, 1::2] = 7

n
```

Создание вертора p:

image.png

```python
# Параметры нормального распределения
math_expectation = 2
dispersion = 4

# Генерация значений из нормального распределения
p = np.random.normal(math_expectation, np.sqrt(dispersion), dimension)

# Прибавляем единицу к каждому элементу массива
p += 1

p
```

Создание вектора q:

image.png

```python
# Вычисление вектора q_t
q_t = n[24, :] * n[71, :]

# Вывод результата
q_t
```

**Индивидуальное задание 1.**

image.png

```python
# Умножение матриц
nm = np.dot(n, m)

# След матрицы nm
trace_nm = np.trace(nm)

# Умножение следа матрицы nm на вектор p
f = np.dot(p, trace_nm)
print(f"f = {f}")
```

**Индивидуальное задание 2.**

image.png

```python
# Умножение числа на вектор
p_2 = np.dot(2, p)

# Транспонирование матрицы
q = q_t.reshape(-1, 1)

# Сумма всех элементов, полученых путем поэлементного умножения элементов с четным i
g = np.sum( p_2[2::2] * q[2::2] )
print(f"g = {g}")
```
